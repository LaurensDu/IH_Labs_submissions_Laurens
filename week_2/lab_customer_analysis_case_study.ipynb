```python
file_1 = pd.read_csv('Data/file1.csv') # reading the docus into the different variables
file_2 = pd.read_csv('Data/file2.csv')
file_3 = pd.read_csv('Data/file3.csv')

file_1.head() # getting an idea of what the table and its columns look like
file_2.head()
file_3.head()
```
